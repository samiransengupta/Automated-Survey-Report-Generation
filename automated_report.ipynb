```python
# generating sample description function

def sample_description(data,column_name):
    results = data[column_name].value_counts(normalize = True).to_frame()*100
    results.loc['Grand Total'] = results.sum(axis = 0)
    results.reset_index(inplace=True)
    results.rename(columns = {'proportion':'Percentage(%)'}, inplace = True)
    results['Percentage(%)'] = results['Percentage(%)'].apply(lambda x: '{0:.2f}'.format(x))
    return results
    
```

```python
gender_wise = pd.crosstab(df['performance_central_gov'],df['gender'],normalize='columns')

gender_wise.sort_values(by='Female',ascending=False, inplace = True)

gender_wise = gender_wise[['Female','Male']]*100

gender_wise.loc['Grand Total'] = gender_wise.sum(axis = 0)

gender_wise[["Female", "Male"]] = gender_wise[["Female", "Male"]].applymap(lambda x: '{:.2f}'.format(x))

gender_wise.reset_index(inplace= True)

gender_wise.rename(columns = {'performance_central_gov':'Opinion','Female':'Female(%)','Male':'Male(%)'}, inplace=True)


gender_wise
```

```python
# generating voting intention fuction

def cross_tab_by_variables(column_1,column_2):
    crosstab_result = pd.crosstab(df[column_1],df[column_2],normalize = 'columns')
    crosstab_result = crosstab_result*100
    crosstab_result.loc['Grand Total'] = crosstab_result.sum(axis=0)
    crosstab_result = crosstab_result.applymap(lambda x: '{:.2f}'.format(x))
    crosstab_result.reset_index(inplace=True)
    
    new_column_names = {column_1:'Opinion'}
    
    for col in crosstab_result.columns[1:]:
        new_column_names[col] = f'{col} (%)'
        
    crosstab_result.rename(columns=new_column_names, inplace=True)
    
    first_percentage_col = crosstab_result.columns[1]

    
    return crosstab_result
```

```python
# gender 

doc.add_paragraph('Chapter 1 â€“ Sample Break up','Heading 1')
doc.add_paragraph('Table 1.1: Gender wise break up', 'Heading 2')

columns = list(gender.columns) # get columns list

table = doc.add_table(rows=1, cols=len(columns),style='Table Grid') # add table with 1 row and len(columns) columns
table.autofit = True # set autofit true

for col in range(len(columns)):
    #set_cell_margins(table.cell(0, col), top=100, start=100, bottom=100, end=50) # set cell margin
    table.cell(0, col).text = columns[col]
    
# iterate on dataframe  

for i, row in enumerate(gender.itertuples()):
    table_row = table.add_row().cells           # add new row to table
    for col in range(len(columns)):             # iterate over each column in row and add text
        table_row[col].text = str(row[col+1]) # avoid index by adding col+1
```

```python
# gender styling
for col in range(len(columns)):
    shading_elm_1 = parse_xml(r'<w:shd {} w:fill="#BEBEBE"/>'.format(nsdecls('w')))
    table.row_cells(0)[col]._tc.get_or_add_tcPr().append(shading_elm_1)
    
last_row_idx = len(table.rows) - 1  # Index of the last row
for col in range(len(columns)):  # Iterate over all columns in the last row
    shading_elm_1 = parse_xml(r'<w:shd {} w:fill="#BEBEBE"/>'.format(nsdecls('w')))
    table.row_cells(last_row_idx)[col]._tc.get_or_add_tcPr().append(shading_elm_1)
    
# Bold & center first row 
for col in range(len(columns)):
    table.cell(0, col).paragraphs[0].runs[0].font.bold = True
    table.cell(0, col).paragraphs[0].alignment = WD_ALIGN_PARAGRAPH.CENTER

# Bold & center last row
for col in range(len(columns)):
    table.cell(len(table.rows) - 1, col).paragraphs[0].runs[0].font.bold = True
    table.cell(len(table.rows) - 1, col).paragraphs[0].alignment = WD_ALIGN_PARAGRAPH.CENTER
    

# Center align all other cells
for row in table.rows[1:-1]:  # Excluding the first and last rows
    for cell in row.cells:
        for paragraph in cell.paragraphs:
            paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER
            
# Left align the strings in the first column
for row in table.rows:
    row.cells[0].paragraphs[0].alignment = WD_ALIGN_PARAGRAPH.LEFT            
```
